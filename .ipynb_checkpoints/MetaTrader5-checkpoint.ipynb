```python
class Broker(object):
    def __init__(self,broker, mt, demo):
        self._broker = broker
        self._demo = demo
        self.mt = mt
        self.password = "Speedfreak@1234"
        self.login = None
        self.ticks = None
        self.tick = None
        self.setBroker(self._broker, self._broker)
        self.frame = None
        
    def setBroker(self,broker,demo):
        self._broker = broker
        self._demo = demo
        if self._broker == "FPMarkets" and self._demo:
            self.login = 773345
            self.server = "FPMarkets-Demo"
        elif self._broker == "FPMarkets" and not self._demo:
            self.login = 8401602
            self.server = "FPMarkets-Live"
        elif self._broker == "xm" and self._demo:
            self.login = 65988020
            self.server = "XMGlobal-MT5 2"
            #return {"login": 65720956, "password" : password, "server": "XMGlobal-MT5 2"}
        elif self._broker == "xm" and not self._demo:
            self.login = 81542102
            self.server = "XMGlobal-MT5 4"
            #return {"login": 84383632, "password" : password,"server": "XMGlobal-MT5 4"}
        elif self._broker == "ICMarkets" and  self._demo:
            self.login = 51030807
            self.server = "ICMarketsSC-Demo"
            #return {"login": 51030807, "password" : "q4S1md6B","server": "ICMarketsSC-Demo"}
        elif self._broker == "ICMarkets" and not self._demo:
            self.login = 84383632
            self.server = "ICMarketsSC-Demo"
            #return {"login": 84383632, "password" : password,"server": "ICMarketsSC-Demo"}

        self.mt.initialize(login = self.login, password = self.password, server = self.server)
        
        print("Initialized")
        
        
    def setTick(self,tick):
        self.tick = tick
    
    def currentTick(self):
        return self.tick 
    
    def currentTicks(self):
        self.ticks = self.mt.symbols_get()
        return self.ticks
    
    def currentBroker(self):
        return self._broker
    
    def setAccountType(self,demo):
        self._demo = demo
    
    def currentAccountType(self):
        if self._demo:
            return "Demo"
        else:
            return "Live"
    
    def checkTick(self):
        if self.tick == None:
            print("Please set tick")
            return True
        return False

    def getTicks(self):
        if self.ticks == None:
            self.currentTicks()
        return [tick._asdict()['name'] for tick in self.ticks]

    def setTimeFrame(self,frame):
        if frame == "H1":
            self.frame = self.mt.TIMEFRAME_H1
        elif frame == "H2":
            self.frame = self.mt.TIMEFRAME_H2
            
        elif frame == "H3":
            self.frame = self.mt.TIMEFRAME_H3
        elif frame == "H4":
            self.frame = self.mt.TIMEFRAME_H4
            
        elif frame == "H6":
            self.frame = self.mt.TIMEFRAME_H6
        elif frame == "M5":
            self.frame = self.mt.TIMEFRAME_M5
            
        elif frame == "M10":
            self.frame = self.mt.TIMEFRAME_M10
            
        elif frame == "M15":
            self.frame = self.mt.TIMEFRAME_M15
        
        elif frame == "M20":
            self.frame = self.mt.TIMEFRAME_M20
            
        elif frame == "M30":
            self.frame = self.mt.TIMEFRAME_M30
        
        elif frame == "D":
            self.frame = self.mt.TIMEFRAME_D1
        
        elif frame == "WK":
            self.frame = self.mt.TIMEFRAME_W1
        
        elif frame == "MN":
            self.frame = self.mt.TIMEFRAME_MN1
            
    def getPastData(self, time_steps):

        if self.checkTick():
            print("nocheck")
            return 
        
        ohlc_data = self.mt.copy_rates_range(self.currentTick(),self.frame, (datetime.now() - timedelta(hours=time_steps)), datetime.now())
        x = pd.DataFrame(ohlc_data)
        print(x.columns)
        x['time'] = pd.to_datetime(x['time'],unit="s")
        x = x.set_index('time')
        return x
    
    def getAllTicksPastData(self,ticks, time_steps):
        #form dict
        all_past_dict = dict()
                    
        #for loop ticks and store in dict the getPastData(tick,frame,time_steps)
        for tick in ticks:
            self.setTick(tick)
            print(tick)
            all_past_dict[tick] = self.getPastData(time_steps)
        
        return all_past_dict
    
    def getInterestingTicks(self, two_point_cut,five_point_cut):
        interesting_ticks = []
        if self.ticks==None:
            self.currentTicks()
            
        for tick in self.ticks:
            self.setTick(tick)
            if tick._asdict()['digits']>2 and tick._asdict()['time']!=0:
                if tick._asdict()['spread']<two_point_cut:
                    interesting_ticks.append(tick.name)
            else:
                if tick._asdict()['spread']<five_point_cut and tick._asdict()['time']!=0:
                    interesting_ticks.append(tick.name)
        return interesting_ticks
    
    def calculateMarginReq(self, action, tick, lot, price): 
        return mt.order_calc_margin(action, tick, lot, price)

    def account_info(self):
        return self.mt.account_info()
    
    def getTick(self):
        return self.mt.symbol_info(self.tick)._asdict()
    
    def getAllinterestMargin_Min(self,ticks,action):
        price = None
        action = None
        all_past_dict = dict()
                    
        #for loop ticks and store in dict the getPastData(tick,frame,time_steps)
        for tick in ticks:
            self.setTick(tick)
            print(tick)
            sym_info = self.mt.symbol_info(tick)._asdict()
            lot = sym_info['volume_min']
            if action == "BUY":
                action = self.mt.ORDER_TYPE_BUY
                price = sym_info['ask']
            else:
                action = self.mt.ORDER_TYPE_SELL
                price = sym_info['bid']
            all_past_dict[tick] = self.calculateMarginReq(action, tick, lot, price)
        return all_past_dict
    
```

```python
import numpy as np
class Instrument(object):
    def __init__(self, ticker, data):
        self.ticker = ticker
        self.dataraw = data
        self.data = self.dataraw.close.to_frame()
        self.LogReturns()
        
    def getData(self):
        return self.data
    
    def getDataCLosePrice(self):
        return self.dataraw.close
    
    def getDataOpenPrice(self):
        return self.dataraw.open
    
    def getDataHighPrice(self):
        return self.dataraw.high
    
    def getDataLowPrice(self):
        return self.dataraw.low
    
    def getDataSpread(self):
        return self.dataraw.spread
    
    def getDataVolume(self):
        return self.dataraw.tick_volume
    
    def LogReturns(self):
        self.data['Log_Returns'] = np.log(self.dataraw.close/self.dataraw.close.shift(1))
      
    def plot_prices(self):
        self.data.close.plot(figsize = (12,8))
        plt.title("Price Chart: {}".format(self.ticker), fontsize = 15)
        
    def plot_returns(self, kind = "ts"):
        if kind == "ts":
            self.data.Log_Returns.plot(figsize = (12,8))
            plt.title("Returns: {}".format(self.ticker), fontsize = 15)
        elif kind == "hist":
            self.data.Log_Returns.hist(figsize = (12,8), bins = int(np.sqrt(len(self.data))))
            plt.title("Frequency of Returns: {}".format(self.ticker), fontsize = 15)
    
    def mean_return(self,freq=None):
        if freq==None:
            return self.data.Log_Returns.mean()
        else:
            resampled_price = self.data.close.resample(freq).last()
            resampled_returns = np.log(resampled_price/resampled_price.shift(1))
            return resampled_returns.mean()
        
    def std_return(self,freq=None):
        if freq==None:
            return self.data.Log_Returns.std()
        else:
            resampled_price = self.data.close.resample(freq).last()
            resampled_returns = np.log(resampled_price/resampled_price.shift(1))
            return resampled_returns.std()
    
    def annualized_perf(self):
        mean_return = round(self.data.Log_Returns.mean() * 252,3)
        risk = round(self.data.Log_Returns.std() * np.sqrt(252),3)
        print("Return: {}| Risk: {}".format(mean_return,risk))
    
    def run_strategy(self, alligator):
        df = self.data.copy()
        df['SMA_S'] = df.close.rolling(alligator[0]).mean()
        df['SMA_L'] = df.close.rolling(alligator[1]).mean()
        #df['SMA_LL'] = df.close.rolling(alligator[2]).mean()
        df.dropna(inplace=True)
        df['position'] = np.nan
        for idx,r in df.iterrows():
            if r.SMA_S>r.SMA_L>r.SMA_LL:
                r['position'] = 1
            elif r.SMA_S<r.SMA_L<r.SMA_LL:
                r['position'] = -1
            else: 
                r['position'] = 0
        df['strategy'] = df.position.shift(1) * df['Log_Returns']
        df["creturns"] = df["Log_Returns"].cumsum().apply(np.exp)
        df["cstrategy"] = df["strategy"].cumsum().apply(np.exp)
        df.dropna(inplace=True)
        perf = df["cstrategy"].iloc[-1] # absolute performance of the strategy
        outperf = perf - df["creturns"].iloc[-1] # out-/underperformance of strategy

        return (np.exp(df['strategy'].sum()),np.exp(df['Log_Returns'].sum()), df)
    
    def test_strategy(self,alligator):
        df = self.data.copy()
        df['SMA_S'] = df.close.rolling(alligator[0]).mean()
        df['SMA_L'] = df.close.rolling(alligator[1]).mean()
        df['SMA_LL'] = df.close.rolling(alligator[2]).mean()
        df.dropna(inplace=True)
        df['position'] = np.nan
        for idx,r in df.iterrows():
            if r.SMA_S>r.SMA_L>r.SMA_LL:
                r['position'] = 1
            elif r.SMA_S<r.SMA_L<r.SMA_LL:
                r['position'] = -1
            else: 
                r['position'] = 0
        df['strategy'] = df.position.shift(1) * df['Log_Returns']
        df["creturns"] = df["Log_Returns"].cumsum().apply(np.exp)
        df["cstrategy"] = df["strategy"].cumsum().apply(np.exp)
        df.dropna(inplace=True)
        perf = df["cstrategy"].iloc[-1] # absolute performance of the strategy
        outperf = perf - df["creturns"].iloc[-1] # out-/underperformance of strategy
        #return (round(perf, 6), round(outperf, 6))
        return -round(perf,6)
   
                
    def optimize_parameters(self, alligator_range_S,alligator_range_L,alligator_range_LL):
        opt = brute(self.test_strategy, (alligator_range_S,alligator_range_L,alligator_range_LL), finish=None)
        return (opt, -self.test_strategy(opt))
```

```python
class EMABacktester(): 
    ''' Class for the vectorized backtesting of EMA-based trading strategies.

    Attributes
    ==========
    symbol: str
        ticker symbol with which to work with
    EMA_S: int
        time window in days for shorter EMA
    EMA_L: int
        time window in days for longer EMA
    start: str
        start date for data retrieval
    end: str
        end date for data retrieval
    tc: float
        proportional transaction costs per trade
        
        
    Methods
    =======
    get_data:
        retrieves and prepares the data
        
    set_parameters:
        sets one or two new EMA parameters
        
    test_strategy:
        runs the backtest for the EMA-based strategy
        
    plot_results:
        plots the performance of the strategy compared to buy and hold
        
    update_and_run:
        updates EMA parameters and returns the negative absolute performance (for minimization algorithm)
        
    optimize_parameters:
        implements a brute force optimization for the two EMA parameters
    '''
    
    def __init__(self, symbol, EMA_S, EMA_L, data, tc):
        self.symbol = symbol
        self.EMA_S = EMA_S
        self.EMA_L = EMA_L
        #self.start = start
        #self.end = end
        self.tc = tc
        self.results = None 
        
        self.data=data
        self.get_data()
    #def __repr__(self):
    #    return "EMABacktester(symbol = {}, EMA_S = {}, EMA_L = {}, start = {}, end = {})".format(self.symbol, self.EMA_S, self.EMA_L, self.start, self.end)
        
    def get_data(self):
        ''' Retrieves and prepares the data.
        '''
        
        #raw = pd.read_csv("forex_pairs.csv", parse_dates = ["Date"], index_col = "Date")
        #raw = raw[self.symbol].to_frame().dropna()
        #raw = raw.loc[self.start:self.end]
        self.data.rename(columns={"close": "price"}, inplace=True)
        self.data["returns"] = np.log(self.data.price/ self.data.price.shift(1))
        self.data["EMA_S"] = self.data["price"].ewm(span = self.EMA_S, min_periods = self.EMA_S).mean() 
        self.data["EMA_L"] = self.data["price"].ewm(span = self.EMA_L, min_periods = self.EMA_L).mean() 
        
    def set_parameters(self, EMA_S = None, EMA_L = None):
        ''' Updates EMA parameters and resp. time series.
        '''
        if EMA_S is not None:
            self.EMA_S = EMA_S
            self.data["EMA_S"] = self.data["price"].ewm(span = self.EMA_S, min_periods = self.EMA_S).mean() 
        if EMA_L is not None:
            self.EMA_L = EMA_L
            self.data["EMA_L"] = self.data["price"].ewm(span = self.EMA_L, min_periods = self.EMA_L).mean()
            
    def test_strategy(self):
        ''' Backtests the trading strategy.
        '''
        data = self.data.copy().dropna()
        data["position"] = np.where(data["EMA_S"] > data["EMA_L"], 1, -1)
        data["strategy"] = data["position"].shift(1) * data["returns"]
        data.dropna(inplace=True)
        
        # determine when a trade takes place
        data["trades"] = data.position.diff().fillna(0).abs()
        
        # subtract transaction costs from return when trade takes place
        data.strategy = data.strategy - data.trades * self.tc
        
        data["creturns"] = data["returns"].cumsum().apply(np.exp)
        data["cstrategy"] = data["strategy"].cumsum().apply(np.exp)
        self.results = data
        
        perf = data["cstrategy"].iloc[-1] # absolute performance of the strategy
        outperf = perf - data["creturns"].iloc[-1] # out-/underperformance of strategy
        return round(perf, 6), round(outperf, 6)
    
    def plot_results(self):
        ''' Plots the cumulative performance of the trading strategy
        compared to buy and hold.
        '''
        if self.results is None:
            print("No results to plot yet. Run a strategy.")
        else:
            title = "{} | EMA_S = {} | EMA_L = {} | TC = {}".format(self.symbol, self.EMA_S, self.EMA_L, self.tc)
            self.results[["creturns", "cstrategy"]].plot(title=title, figsize=(12, 8))
        
    def update_and_run(self, EMA):
        ''' Updates EMA parameters and returns the negative absolute performance (for minimization algorithm).

        Parameters
        ==========
        EMA: tuple
            EMA parameter tuple
        '''
        self.set_parameters(int(EMA[0]), int(EMA[1]))
        return -self.test_strategy()[0]
    
    def optimize_parameters(self, EMA1_range, EMA2_range):
        ''' Finds global maximum given the EMA parameter ranges.

        Parameters
        ==========
        EMA1_range, EMA2_range: tuple
            tuples of the form (start, end, step size)
        '''
        opt = brute(self.update_and_run, (EMA1_range, EMA2_range), finish=None)
        return opt, -self.update_and_run(opt)
    
```

```python
def test_strategy(df, alligator):
    df = df.copy()
    df['SMA_S'] = df.close.rolling(alligator[0]).mean()
    df['SMA_L'] = df.close.rolling(alligator[1]).mean()
    df['SMA_LL'] = df.close.rolling(alligator[2]).mean()
    df.dropna(inplace=True)
    df['position'] = np.nan
    for idx,r in df.iterrows():
        if r.SMA_S>r.SMA_L>r.SMA_LL:
            r['position'] = 1
        elif r.SMA_S<r.SMA_L<r.SMA_LL:
            r['position'] = -1
        else: 
            r['position'] = 0
    df['strategy'] = df.position.shift(1) * df['Log_Returns']
    df.dropna(inplace=True)
    
    return np.exp(df['strategy'].sum())
```

```python
import MetaTrader5 as mt5

#login = account_dict['login']
#password = account_dict['password']
#server = account_dict['server']

symbol = "GOLD#"
lot = 0.01
point = mt5.symbol_info(symbol).point
price = mt5.symbol_info_tick(symbol).ask
deviation = 20
request = {
    "action": mt5.TRADE_ACTION_DEAL,
    "symbol": symbol,
    "volume": lot,
    "type": mt5.ORDER_TYPE_SELL,
    "price": price,
    "deviation": deviation,
    "magic": 234000,
    "comment": "python script open",
    "type_time": mt5.ORDER_TIME_GTC,
    "type_filling": mt5.ORDER_FILLING_IOC,
}
 
# send a trading request
result = mt5.order_send(request)
# check the execution result
print("1. order_send(): by {} {} lots at {} with deviation={} points".format(symbol,lot,price,deviation));
if result.retcode != mt5.TRADE_RETCODE_DONE:
    print("2. order_send failed, retcode={}".format(result.retcode))
    # request the result as a dictionary and display it element by element
    result_dict=result._asdict()
    for field in result_dict.keys():
        print("   {}={}".format(field,result_dict[field]))
        # if this is a trading request structure, display it element by element as well
        if field=="request":
            traderequest_dict=result_dict[field]._asdict()
            for tradereq_filed in traderequest_dict:
                print("       traderequest: {}={}".format(tradereq_filed,traderequest_dict[tradereq_filed]))
    print("shutdown() and quit")

print("2. order_send done, ", result)
print("   opened position with POSITION_TICKET={}".format(result.order))
print("   sleep 2 seconds before closing position #{}".format(result.order))
time.sleep(2)
```
